#**Data** **Cleaning**

```python
#handling duplicates
df.duplicated().sum()
```

```python
#drop the duplicates
df.drop_duplicates(inplace=True)
```

```python
#Clean column names
df.columns = (
    df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace(r"[^\w_]", "", regex=True)
)
```

```python
#Handle missing values
#Numerical → fill with median
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
#Categorical → fill with mode (most common)
cat_cols = df.select_dtypes(include="object").columns
for col in cat_cols:
    if not df[col].mode().empty:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#after handling missing values
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include='number').columns
correlation_matrix = df[num_cols].corr()
print(correlation_matrix)
```

Pearson correlation matrix using the corr() function to understand the strength and direction of relationships between obesity, physical inactivity, and other numeric attributes.

```python
#merging the inactivity and obesity
obesity_df = df[df['topic'] == 'obesity / weight status']
activity_df = df[df['topic'] == 'physical activity - behavior']
avg_obesity_by_state = obesity_df.groupby('locationdesc')['data_value'].mean().sort_values(ascending=False).reset_index()
avg_obesity_by_state = avg_obesity_by_state.rename(columns={'data_value': 'average_obesity'})
avg_inactivity_by_state = activity_df.groupby('locationdesc')['data_value'].mean().reset_index()
avg_inactivity_by_state = avg_inactivity_by_state.rename(columns={'data_value': 'average_inactivity'})
merged_df = pd.merge(avg_obesity_by_state, avg_inactivity_by_state, on='locationdesc', how='inner')
print(merged_df.head())
```

At first ,separated obesity and physical inactivity data using topic filtering. Then I calculated state-wise averages for both variables. After that, I merged the two datasets using location as the common key. This allowed me to directly compare average obesity and inactivity levels for each state.I used inner join because my analysis required both obesity and inactivity values to be present.

```python
#after changing column names
df.columns
```

```python
#Remove extreme outliers (IQR method) — numeric only
for col in num_cols:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    low = Q1 - 1.5 * IQR
    high = Q3 + 1.5 * IQR
    df = df[(df[col] >= low) & (df[col] <= high)]
print("IQR:", IQR)
print("Lower:", low)
print("Upper:", high)
print("Outliers:\n", df)
```

To remove extreme values, I applied the IQR method. For each numerical column, I calculated the 25th and 75th percentiles, computed the interquartile range, and removed observations falling outside 1.5 times the IQR. This ensures robust outlier removal without assuming normal distribution

# **Statistical Analysis**

```python
#Statistical Analysis
statistical_cols=['data_value', 'data_value_alt', 'low_confidence_limit', 'high_confidence_limit', 'sample_size']
mean = df[statistical_cols].mean()
median = df[statistical_cols].median()
mode = df[statistical_cols].mode()
std = df[statistical_cols].std()
var = df[statistical_cols].var()
percentile = df[statistical_cols].quantile([0.25, 0.5, 0.75])
skewness = df[statistical_cols].skew()
kurtosis = df[statistical_cols].kurt()

print("\nmean :\n",mean)
print("\nmedian :\n",median)
print("\nmode :\n",mode)
print("\nstandard deviation :\n",std)
print("\nvariance :\n",var)
print("\npercentile :\n",percentile)
print("\nSkewness:\n", skewness)
print("\nKurtosis:\n", kurtosis)
```

## **Visualize Measures of Central Tendency**

```python
#importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
central_tendency_df = pd.DataFrame({'Mean': mean, 'Median': median}).T
display(central_tendency_df)
```

```python
# Iterate through each numerical column to create boxplots
# Calculate central tendency measures
mean_values = df[statistical_cols].mean()
median_values = df[statistical_cols].median()
mode_values = df[statistical_cols].mode().iloc[0]

# Grid size
n_cols = 2
n_rows = int(np.ceil(len(statistical_cols) / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 5*n_rows))
axes = axes.flatten()

for i, col in enumerate(statistical_cols):
    sns.boxplot(y=df[col], ax=axes[i])
    axes[i].set_title(f'Box Plot - {col}')
    axes[i].set_ylabel(col)

# Remove empty plots if any
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

computed the mean, median, and mode to understand the central tendency of numerical variables. Then I created boxplots in a dynamic grid layout to visualize distribution, detect skewness, and identify extreme outliers using the IQR method.
